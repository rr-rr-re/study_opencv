```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("/work/data/src/grapes.jpg", 0)
display(Image.fromarray(img))
```

```python
threshold = 100
ret, img_th = cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY)
```

```python
ret
```

```python
display(Image.fromarray(img_th))
```

```python
ret2, img_o = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
ret2
```

```python
display(Image.fromarray(img_o))
```

```python
img_ada = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 1)
```

```python
display(Image.fromarray(img_ada))
```

```python
img_f = cv2.imread("/work/data/src/floor.jpg", 0)
display(Image.fromarray(img_f))
```

```python
def onTrackbar(position):
    global threshold
    threshold = position
cv2.namedWindow("img_f")
threshold = 100
cv2.createTrackbar("track", "img", threshold, 255, onTrackbar)
while True:
    ret, img_th1= cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY)
    display(Image.fromarray(img_th1))
    if cv2.waitKey(10) == 27:
        break
cv2.destroyAllwindows
```
