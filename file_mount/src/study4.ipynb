```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ganma = 1.5
img = cv2.imread("/work/data/src/Berry.jpg")
ganma_cvt =  np.zeros((256, 1), dtype=np.uint8)
for i in range(256):
    ganma_cvt[i][0] = 256 * (float(i)/255) ** (1.0 / ganma)
    
img_ganma = cv2.LUT(img, ganma_cvt)

display(Image.fromarray(img))
display(Image.fromarray(img_ganma))
```
