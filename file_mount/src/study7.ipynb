```python
import cv2
from PIL import Image
import numpy as np
```

```python
img = cv2.imread("/work/data/src/grapes.jpg")
h, w = img.shape[:2]
dx, dy = 30, 30
```

```python
afn_mat = np.float32([[1, 0, dx], [0, 1, dy]])
img_afn = cv2.warpAffine(img, afn_mat, (h, w))
```

```python
display(Image.fromarray(img_afn))
```
